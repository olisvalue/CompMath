### В данной практической работе займёмся численным интегрированием. 

В работе предусмотрим автоматическое изменение кол-ва числа узлов для получения назначенной точности. Применим правило Рунге для оценки погрешности.

Рассмотрим следующие вопросы:

3) Когда отсутствует даже 1 производная в нуле.

Сравним с точным ответом для контроля результатов

Что будем делать:  
1) Посчитаем несобственный интеграл   с точностью $10^{-4}$

2) Посчитаем интеграл негладкой функции $|x|$

Протестируем два вида сгущающихся сеток:  
а) особенность в производной попадает в узел  
б) не попадает

Посчитаем по формуле трапеций
Скорость сходимость и погрешность оценим по правилу Рунге
 

#### Формула трапеций:
Ожидаем порядок точности составной формулы = 2, так как на одном отрезке формула трапеций имеет порядок точности 3

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 1/((1+x)*np.sqrt(x))

#для Функции Рунге отлично работает (нет особенностей)
#def f(x):
    #return 1/(1 + x**2)

N = 20 #число сгущений
err = np.zeros(N)
I_tr = np.zeros(N)

a, b = 0.001, 1 #задаём границы вычисления интеграла
n = 2**np.arange(N) #числа разбиений
h = (b-a)/n #шаги разбиения

#I_ex = np.pi/2
#на отрезке 0, 1 этот интеграл вычисляется как pi/2
I_ex = 2*(np.arctan(np.sqrt(b)) - np.arctan(np.sqrt(a))) #точное значение интеграла на отрезке

#Отрисуем интегрируемую функцию
points = np.linspace(a, b, 10000)
plt.plot(points, f(points), '-g')
plt.grid(True)
plt.show()


eps = 10**-6 #желаемая ошибка

for k in range(N):
    x = np.linspace(a, b, n[k] + 1)
    fval = f(x)
    #I_tr - приближённое значение интеграла
    I_tr[k] = h[k] * (np.sum(fval) - 0.5 * (fval[0] + fval[-1]))
    err[k] = np.abs(I_tr[k] - I_ex)
    #критерий остановки (правило Рунге:)
    #if (I_tr[k] - I_tr[k-1])/(2**n[k] - 1) < eps and h[k] <= 10**-3:
    if (abs(I_tr[k] - I_tr[k-1]))/(3) < eps and h[k] <= 10**-5:
        print("На этой итерации, по правилу Рунге, достигнута желаемая точность")
    print(I_tr[k], I_ex)
fig, ax = plt.subplots()

#Проверим, что ошибка меняется как O(h^2)
plt.loglog(h, err, 'ko', label = 'Ошибка')
plt.loglog(h, err[-1] * (h/h[-1])**2 , 'r-', label = '$O(h^2)$ - 2й порядок')
plt.legend()
plt.grid(True)
plt.show()
print('errors:', err)
```

### Посчитаем интеграл по формуле Гаусса на (0,1)

```python
from scipy import integrate

def f(x):
    return 1/((1+x)*np.sqrt(x))

I_ex = np.pi/2

Ig, a = integrate.fixed_quad(f, 0, 1, n = 1000)

print('точное знач.', I_ex)
print('знач. по ф-ле Гаусса: ', Ig)
print('Ошибка: ', abs(I_ex - Ig))
```

### Вернёмся к поставленным вопросам и дадим на них ответ

1) Как особенность в нуле влияет на алгоритм применения правила Рунге?

    -Как показало вычисление интеграла с особенностью в нуле, если отграничить её, то мы вычисляем интеграл точно, и правило Рунге свидетельствует об этом. Когда область "приближается" к особенности (как у нас, (0.00000001, 1)), то точность, конечно, падает, но и правило Рунге перестаёт сообщать о том, что получена нужная точность.
    -Проще говоря, особенность в нуле не повлияла на алгоритм применения правила Рунге, оно всё так же верно свидетельствует о том, достигнута заданная точность или же нет
2) Какова фактическая точность формулы Симпсона в этом случае

    -Фактическая точность формулы Симпсона, или трапеций, как и любой другой составной, сильно падает, если приближаться к особенности функции. По понятным причинам: в оценке ошибки квадратурной формулы присутствует оценка на максимальную производную в области, а в окрестности особой точки производная бесконечна - значит, составная формула может дать любой результат.
3) Когда отсутствует даже 1 производная в нуле.

    -Если отсутствует даже 1-я производная в нуле, то в окрестности этой точки составная формула не гарантирует точного результата - для точной оценки требуется гладкость функции в точке и знание производной.

Производные (и старших порядков) у подобных функций с особенностями большие в окрестности особой точки (здесь - нуля), но её можно отграничить.

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 1/((1+x)*np.sqrt(x))
def f_eps(x):
    return x**(-1/2) - x **(1/2) + x**(3/2) - x**(5/2) + x**(7/2)

def integr_eps(x):
    return 2 * x**(1/2) -2/3*x**(3/2) + 2/5 * x**(5/2) - 2/7 * x**(7/2) + 2/9 * x **(9/2)

#для Функции Рунге отлично работает (нет особенностей)
#def f(x):
    #return 1/(1 + x**2)

N = 20 #число сгущений
err = np.zeros(N)
I_tr = np.zeros(N)

a, b = 0.01, 1 #задаём границы вычисления интеграла
n = 2**np.arange(N) #числа разбиений
h = (b-a)/n #шаги разбиения

I_ex = np.pi/2
eps = 10**-6 #желаемая ошибка

I_g = integr_eps(a)

for k in range(N):
    x = np.linspace(a, b, n[k] + 1)
    fval = f(x)
    #I_tr - приближённое значение интеграла
    I_tr[k] = h[k] * (np.sum(fval) - 0.5 * (fval[0] + fval[-1])) + I_g
    err[k] = np.abs(I_tr[k] - I_ex)
    #критерий остановки (правило Рунге:)
    #if (I_tr[k] - I_tr[k-1])/(2**n[k] - 1) < eps and h[k] <= 10**-3:
    if (abs(I_tr[k] - I_tr[k-1]))/(3) < eps and h[k] <= 10**-5:
        print("На этой итерации, по правилу Рунге, достигнута желаемая точность")
    print(I_tr[k], I_ex)

#Проверим, что ошибка меняется как O(h^2)
plt.loglog(h, err, 'ko', label = 'Ошибка')
plt.loglog(h, err[-1] * (h/h[-1])**2 , 'r-', label = '$O(h^2)$ - 2й порядок')
plt.legend()
plt.grid(True)
plt.show()
print('errors:', err)
```
