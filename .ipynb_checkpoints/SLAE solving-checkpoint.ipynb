### Создадим матрицу А. Найдём её число обусловленности. Оценим априорно погрешность полученного решения. (Используется норма модуля макс. элемента)


1. Наименьший вычисляемый элемент в матрице А, не считая нуля, равняется 0.5. При этом ошибка округления для каждого числа составляет порядка $\epsilon_{маш}$, равный $10^{-7}$ в случае np.float и $10^{-16}$ в случае np.double. Стало быть, $ \frac{||\Delta A||}{|| A||}$ - пренебрежимо мал, порядка 10^{-16}. Формула оценки ошибки решения ур-я Ax = b упрощается в следующую: $\frac{||\Delta x||}{||x||} <= cond(A) \cdot \frac{||\Delta b||}{||b||}$ 
2. Чтобы оценить априорно погрешность решения, необходимо знать "точное" решение. Как же его узнать? Зададим его сами, как вектор $x_{точноe} = (1, 1, ..., 1)$ и векторно умножим на матрицу A, получив таким образом вектор b, соответствующий данному решению. Далее, решим полученную систему Ax = b относительно неизвестного x, и сравним полученное $x_{точное}$ с $x_{реш}$.
Решили, что $||\Delta b||$ определяется машинной точностью округления, т.е равен порядка $||10^{-16} \cdot b||_1$  (по норме максимума). Стало быть, $\frac{||\Delta b||_1}{||b||_1 = 600}$ в данной оценке будет порядка $10^{-16}$ и всё сводится к величине числа обусловленности матрицы A. Рассчитаем его при помощи библиотек Numpy:

Итак, по норме 1 число обусловленности матрицы A равняется 21. Следовательно, ошибка точного решения априорно $\frac{||\Delta x||}{||x||} <= 21 \cdot 10^{-16} \simeq 2.1 \cdot 10^{-15} $    
Проверим это, запустив соответствующий код:

```python
n = 100
a = 0
matrix_A = np.zeros((n, n), dtype = np.double) #создадим матрицу A
vector_f = np.arange(1, n + 1) #создадим вектор f

for i in range(n):
    for j in range(n):
        if (i == j):
            matrix_A[i][j] = a
        elif (j >= i + 2):
            matrix_A[i][j] = 0
        else:
            matrix_A[i][j] = (i + 1)/(j + 1)
```

```python
cond_A = np.linalg.cond(matrix_A, p = np.inf)
cond_A #Нашли число обусловленности матрицы А по 1-норме
```

```python
x = 1 * np.ones(n, dtype = np.double)
print(np.linalg.norm(x, 1))
#print(x)
b = matrix_A @ x 
#print(b)
y = np.linalg.solve(matrix_A, b) #поскольку работаем по норме 1(модуль макс. элем.), и x = (1, 1, .. 1)
#то найденная абсолютная ошибка по сути равна относительной в этой норме (делим на единицу)
print('Ошибка: = {0:6.3e}'.format(np.amax(np.abs(y - x))))
print(b)
```

Интересно, что фактическая ошибка оказалась равной максимальной априорно оцененной (в 1-ой норме). Получается, в данной матрице, выбирая вектор b, соответствующий вектору решения (1, 1... 1), случайно получили вектор, соответствующий максимальной ошибке.  
Для np.float(23 бита под мантиссу, и маш. эпсилон = 10^{-7});   Ошибка: = 8.345e-07  
Для np.double(52 бита под мантиссу, и маш. эпсилон = 10^{-16});  Ошибка: = 2.109e-15  

### А как испортить обсуловленность этой матрицы? Достаточно поставить a = 0, и тогда ошибка (по первой норме) возрастает до 1.667e-01 в double. И какой подобрать столбец f, чтобы усилить эффект?

В соответствии с выражением $min \dfrac{\dfrac{||\Delta x||}{||x||}}{\dfrac{||\Delta b||}{||b||}} >= \dfrac{1}{max \dfrac{||\Delta b||}{||\Delta x||}} \cdot \dfrac{1}{max \dfrac{||x||}{||b||}}$ Нужно подбирать столбец так, чтобы ||f|| >> ||x||  
В нашем случае, например, работает следующее: пусть решение x = ($10^6$, $10^6$, ..., $10^6$). Тогда ошибка в double составляет 2.561e-09. Но происходит это из - за относительности машинной погрешности. Чтобы подобрать столбец f математически хитро, надо выбрать его так, чтобы достигался max $\dfrac{||x|| = ||A^{-1} b||}{||b||}$, т.е вектор b надо выбрать (по норме 1) так, чтобы строчка матрицы A^{-1} с максимальной суммой модулей умножалась на $(sign(a_{i0}), sign(a_{i1}) ...)$. 

### Получим точное решение системы, используя метод Гаусса в библиотеке np.linalg

```python
sol = np.linalg.solve(matrix_A, vector_f)
sol #Получили "точное" решение системы, используя np.linalg.solve (метод Гаусса)
```

## Реализуем метод простой итерации для решения этой системы. 

$$x_{k+1} = (E - \tau  A)x_k - \tau b $$

Проведём априорную оценку сходимости метода Зейделя (при оптимальном параметре tau). Предполагая, что решение системы нам неизвестно.  
Будем оценивать по следующей формуле, имеющей место во всех "сжимающих" итерационных методах: $$ ||e_k|| <= \dfrac{||S||^k}{1 - ||S||} * ||x_1 - x_0|| $$ 
Найдём оптимальный параметр $\tau^*$, как $\tau^*$ = $\dfrac{2}{|\lambda_{max}| + |\lambda_{min}|}$  
Для этого найдём собственные числа:

```python
eigvals = np.linalg.eigvals(matrix_A) #Нашли собственные значения матрицы А
print(np.abs(eigvals)) #модули собств. чисел
np.linalg.norm(matrix_A, 2)
```

```python
minmodule = np.abs(np.amin(eigvals))
maxmodule = np.abs(np.amax(eigvals))
print('Минимальный модуль =', minmodule, 'Максимальный модуль = ', maxmodule)
tau_opt = 2 / (minmodule + maxmodule)
print('Границы выбора Тау для сходимости метода: 0 < tau <', 2 / maxmodule)
print('Оптимальное Тау: ', tau_opt)
```

```python
matrix_E = np.diagflat(np.ones(100)) #создали единичную матрицу
np.linalg.norm(matrix_E - (tau_opt) * matrix_A, 2)
```

Итак, oптимальное $\tau ^*$: $\approx 0.01$   
Реализуем метод простой итерации:

```python
tau = tau_opt#выбрали оптимальное тау
matrix_E = np.diagflat(np.ones(100)) #создали единичную матрицу
k = 20#число итераций
x_0 = np.zeros(100, dtype = np.single) #нулевой вектор - начальный
x_actual = x_0 
sigma = np.zeros(k)
difference_x = np.zeros(k)
residual_vec = np.zeros(k)
for i in range(k):
    prev = x_actual
    residual_vec[i] = np.linalg.norm(matrix_A @ x_actual - vector_f, np.inf) 
    x_actual = (matrix_E - tau*matrix_A) @ x_actual + tau * vector_f
    difference_x[i] = np.linalg.norm(x_actual - prev, np.inf) 
    sigma[i] = np.linalg.norm(x_actual - sol, np.inf)
```

```python
print(' отклонение от точного ||x_i - x_{точное}||\n', sigma) #то есть отклонение от точного ||x_i - x_{точное}||
```

```python
print('||x_i - x_{предыдущее}||\n',difference_x) #то есть ||x_i - x_{предыдущее}||
```

```python
print('Невязка по норме Max Element\n', residual_vec) #то есть невязка
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
x = np.arange(1, k + 1)
plt.yscale('log')
plt.plot(x, sigma, label = '$||x_i - x_{точное}||_1$', color = np.random.rand(3))
plt.plot(x, difference_x, label = '$||x_i - x_{i - 1}||_1$', color = np.random.rand(3))
plt.plot(x, residual_vec, label = 'Невязка', color = np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("k-iteration")
plt.ylabel("")
plt.title("Statistics of simple ither method with MAX-norm")
plt.grid()

plt.show()
```

## Но ведь матрица не подходила для метода простой итерации! (матрица не симметрична) И метод сошёлся, можно сказать, лишь по стечению случая. Исправим это, выполнив симметризацию матрицы.

Наглядная симметризация простой матрицы:

```python
matrix_one = np.array([[1,3, 20], [2, 5, 17], [1, -23, 0]])
print('first matrix:\n', matrix_one)
matrix_two = matrix_one.transpose()
print('second matrix:\n', matrix_two)
print('multiplied:\n',matrix_one @ matrix_two)
```

Видно, что значения по модулю у получившейся матрицы сильно возрастают. Аналогичное произошло и с нашей матрицей

```python
symm_matrix_A = matrix_A.transpose() @ matrix_A 
symm_vector_f = matrix_A.transpose() @ vector_f 
```

```python
matrix_A = symm_matrix_A
vector_f = symm_vector_f
eigvals = np.linalg.eigvals(symm_matrix_A) #Нашли собственные значения матрицы А
print('eigvals:\n',(eigvals))
print('eucildean norm of symm_matrix_A:\n',np.linalg.norm(matrix_A, 2))
```

```python
matrix_E = np.diagflat(np.ones(100)) #создали единичную матрицу
opt_q = np.linalg.norm(matrix_E - (tau_opt) * symm_matrix_A, 2)
print(opt_q)
```

Довольно низкая скорость сходимости!

Воспользуемся следующей оценкой, чтобы оценить кол-во итераций (допустим, хотим точность $10^{-6}$): $$ ||e_k|| <= \dfrac{||S||^k}{1 - ||S||} * ||x_1 - x_0|| $$ 

```python
tau = tau_opt#выбрали оптимальное тау
matrix_E = np.diagflat(np.ones(100)) #создали единичную матрицу
x_0 = np.zeros(100, dtype = np.single) #нулевой вектор - начальный
x_1 = (matrix_E - tau*matrix_A) @ x_0 + tau * vector_f
first_err = np.amax(x_1 - x_0) ## x_1 - x_0 по норме макс. элемента
print(first_err)
```

```python
import math
accuracy = 10**-6
num = 10**-6 * (1 - opt_q) / first_err
num_iterations = math.log(num, opt_q)
print('Нужно не больше {} итераций для точности {} по норме макс. элемента'.format(num_iterations, accuracy))
```
