```python
import numpy as np
```

```python
H = np.array([8, 10, 15, 20, 30, 40, 60, 80])
B = np.array([13, 14, 15.4, 16.3, 17.2, 17.8, 18.5, 18.8])
ones = np.ones(H.size)
```

```python
m00 = ones @ ones
m01 = ones @ H
m10 = H @ ones
m11 = H @ H
M = np.matrix([[m00, m01], [m10, m11]])
print('M matrix: ')
print(M)
f = H/B
b = np.array([f @ ones, f @ H])
print('b vector:', b)
M_obr = np.linalg.inv(M)
print('Inv M matrix: ')
print(M_obr)
A = M_obr @  b
print('coefs: ', A)
```

```python
A = np.array(A)
np.resize(A, (0,2) )
a = A[0][0]
b = A[0][1]
def foo_B(H):
    return H/(a + b*H)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 5))
plt.plot(H, B, label = 'points', color = np.random.rand(3))

H_approx = np.arange(5, 85)
B_approx = foo_B(H_approx)

plt.plot(H_approx, B_approx, label = 'approx', color = np.random.rand(3))
plt.legend(loc="upper right", bbox_to_anchor=(1.4, 0.95))
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Approximation with MNK")
plt.grid()

plt.show()
```
